##  Python for Loop Questions

### 1. Write a Python program to print numbers from 1 to 50 using a for loop.

```python
for i in range(1, 51):
    print(i)
```

2. ### Print all even numbers between 1 and 100 using a for loop.

```python
for i in range(2, 101, 2):
    print(i)
```

### 3. Write a program to calculate the sum of numbers from 1 to N entered by the user.

```python
n = int(input("Enter a number: "))
total = 0

for i in range(1, n + 1):
    total += i

print("Sum =", total)
```

### 4. Print the multiplication table of a number entered by the user (up to 10).

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

### 5. Write a program to reverse a string using a for loop.


```python
text = input("Enter a string: ")
reverse = ""

for char in text:
    reverse = char + reverse

print("Reversed string:", reverse)
```

### 6. Count the number of vowels present in a string using a for loop.

```python
text = input("Enter a string: ")
count = 0

for char in text:
    if char in "aeiouAEIOU":
        count += 1

print("Number of vowels:", count)
```

### 7. Print the following pattern:

### 8. Write a program to check whether a number is prime using a for loop.

```python
num = int(input("Enter a number: "))
prime = True

if num <= 1:
    prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            prime = False
            break

if prime:
    print("Prime number")
else:
    print("Not a prime number")
```

### 9. Given a list:
numbers = [12, 5, 7, 18, 21, 30]
Print only numbers divisible by 3.

```python
numbers = [12, 5, 7, 18, 21, 30]

for num in numbers:
    if num % 3 == 0:
        print(num)
```

### 10. Break and Continue
Print numbers from 1 to 20 but: - Skip multiples of 3. - Stop execution if number becomes greater than 15.

```python
for i in range(1, 21):
    if i > 15:
        break
    if i % 3 == 0:
        continue
    print(i)
```

### 12. Calculate factorial of a number using a while loop.

```python
num = int(input("Enter a number: "))
fact = 1
i = 1

while i <= num:
    fact *= i
    i += 1

print("Factorial is:", fact)
```
