## 1. Student Grade Calculator
Write a function calculate_grade(marks) that takes marks as input and returns grade according to:

90 and above â†’ A
80 -- 89 â†’ B
70 -- 79 â†’ C
Below 70 â†’ Fail


```python
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "Fail"

# Taking input from user
marks = int(input("Enter your marks: "))

# Calling the function
grade = calculate_grade(marks)

# Printing the result
print("Your grade is:", grade)
```

## 2. Even or Odd Number Checker
Write a function is_even(number) that returns:

True if the number is even
False if the number is odd

```python
def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False

print(is_even(10))   # Output: True
print(is_even(7))    # Output: False
```

## 3. Factorial Using Function
Write a function factorial(n) that returns the factorial of a number using a loop.

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result = result * i
    return result

print(factorial(5))  
```

## 4. Find Maximum Number
Write a function find_max(a, b, c) that returns the largest number among three numbers.

```python
def find_max(a, b, c):
    return max(a, b, c)

print(find_max(10, 25, 15))  # Output: 25
```

## 5. Word Counter
Write a function count_words(sentence) that counts how many words exist in a given sentence.

```python
def count_words(sentence):
    words = sentence.split()
    return len(words)

print(count_words("I love Python programming"))  # Output: 4
```

## 6. Prime Number Checker
Write a function is_prime(n) that checks whether a number is a prime number.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(7))   # Output: True
print(is_prime(10))  # Output: False
```

## 7. List Sum Function
Write a function list_sum(numbers) that takes a list as input and returns the sum of all numbers.

```python
def list_sum(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(list_sum([10, 20, 30]))  # Output: 60
```

## 8. Password Strength Checker
Write a function check_password(password) that checks:

Minimum 8 characters required
Must contain at least one digit

```python
def check_password(password):
    if len(password) >= 8 and any(char.isdigit() for char in password):
        return "Strong"
    else:
        return "Weak"

print(check_password("pass1234"))  # Output: Strong
print(check_password("hello"))     # Output: Weak
```

## 9. Temperature Converter
Write a function celsius_to_fahrenheit(c) that converts temperature from Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

print(celsius_to_fahrenheit(0))   # Output: 32.0
```

## 10. Remove Duplicates from List
Write a function remove_duplicates(data) that removes duplicate elements from a list and returns a new list.

```python
def remove_duplicates(data):
    return list(set(data))

print(remove_duplicates([1, 2, 2, 3, 4, 4]))  # Output: [1, 2, 3, 4]
```
