```python
# -------------------------------
# 1. Buying an item with tax
# -------------------------------

money = 50
item_price = 15
tax_rate = 0.03

tax = item_price * tax_rate
total_cost = item_price + tax
money_left = money - total_cost

print("Money left after purchase: $", round(money_left, 2))
```

```python
### -------------------------------
# 2. Days until birthday
# -------------------------------

days = int(input("How many days until your birthday? "))
weeks = days / 7

print("Weeks until your birthday:", weeks)
```

```python

# -------------------------------
# 3. Letter grade program
# -------------------------------

Grade = int(input("Enter your Grades ")) # change this number to test different grades

if 90 <= Grade <= 100:
    print("Grade: A")
elif 80 <= Grade <= 89:
    print("Grade: B")
elif 70 <= Grade <= 79:
    print("Grade: C")
elif 60 <= Grade <= 69:
    print("Grade: D")
elif 0 <= Grade <= 59:
    print("Grade: F")
else:
    print("Invalid grade")
```
