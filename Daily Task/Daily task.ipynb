```python
# Create two sets
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# add()
set_a.add(10)
print("After add():", set_a)

# copy()
set_c = set_a.copy()
print("Copy of set_a:", set_c)

# discard()
set_a.discard(2)   # removes 2 if it exists (no error if not)
print("After discard():", set_a)

# union()
union_set = set_a.union(set_b)
print("Union of sets:", union_set)

# intersection()
intersection_set = set_a.intersection(set_b)
print("Intersection of sets:", intersection_set)

# difference()
difference_set = set_a.difference(set_b)
print("Difference (set_a - set_b):", difference_set)

# clear()
set_a.clear()
print("After clear():", set_a)
```
