## String methods

Print the user’s name in all uppercase letters.
Print the user’s name in all lowercase letters.
Print the length of the user’s name (number of characters).
Replace all the spaces in the user’s name with hyphens (-), and print the result.
Bonus: Trim any leading or trailing spaces (if any) from the user’s name using strip().

```python
# Prompt the user to enter their full name
full_name = input("Please enter your full name: ")

# Bonus: Trim leading and trailing spaces
full_name = full_name.strip()

# Print the name in uppercase
print("Uppercase:", full_name.upper())

# Print the name in lowercase
print("Lowercase:", full_name.lower())

# Print the length of the name
print("Length:", len(full_name))

# Replace spaces with hyphens
print("With hyphens:", full_name.replace(" ", "-"))
```

## Type casting and numeric operations

Ask the user to enter their birth year (as input). Convert that input to an integer.
Calculate the user’s approximate age by subtracting the birth year from the current year. Print the result in a sentence.
Example output:
You are approximately 25 years old.

```python
# Ask the user to enter their birth year
birth_year = input("Enter your birth year: ")

# Convert the input to an integer
birth_year = int(birth_year)

# Get the current year
from datetime import datetime
current_year = datetime.now().year

# Calculate approximate age
age = current_year - birth_year

# Print the result
print(f"You are approximately {age} years old.")
```

### Part 2: Lists and Indexing



Create a list from user input

Prompt the user to enter three of their favorite fruits (e.g., "apple,banana,orange") in one line, separated by commas.
Split the string by commas to create a list called fruit_list.
Print the list to verify that it’s correct.

```python
# Prompt the user to enter three favorite fruits
user_input = input("Enter three of your favorite fruits, separated by commas: ")

# Split the input string into a list
fruit_list = user_input.split(",")

# Optional: remove extra spaces around each fruit
fruit_list = [fruit.strip() for fruit in fruit_list]

# Print the list to verify
print(fruit_list)
```

## List indexing and slicing

Print the first fruit in the list.
Print the last fruit in the list.
Print the second fruit in the list using indexing (i.e., fruit_list[1]).
Print the slice of the list containing the first two fruits.

```python
fruit_list = ['apple', 'banana', 'orange']

# Print the first fruit
print(fruit_list[0])

# Print the last fruit
print(fruit_list[-1])

# Print the second fruit using indexing
print(fruit_list[1])

# Print the slice containing the first two fruits
print(fruit_list[0:2])
```

## List methods

Append a new fruit to the list using append(). Print the updated list.
Remove one fruit from the list using remove() (or pop()). Print the updated list again.
Print the length of the list using len().

```python
fruit_list = ['apple', 'banana', 'orange']

# Append a new fruit
fruit_list.append('mango')
print(fruit_list)

# Remove one fruit (using remove)
fruit_list.remove('banana')
# OR you could use: fruit_list.pop()  # removes the last item
print(fruit_list)

# Print the length of the list
print(len(fruit_list))
```

## Part 3 (Optional Challenge): Combining Concepts

String & List Interaction

Create a new list from the letters of your full_name (excluding spaces).

Hint:

letters_list = list(full_name.replace(" ", ""))
Print the first 3 letters and the last 3 letters of letters_list.

```python
# -----------------------------
# String & List Interaction
# -----------------------------

# Define full name
full_name = "Nabeel"

# Create a list of letters (excluding spaces)
letters_list = list(full_name.replace(" ", ""))

# Print first 3 and last 3 letters
print("First 3 letters:", letters_list[:3])
print("Last 3 letters:", letters_list[-3:])

# -----------------------------
# Manipulate and Reconstruct
# -----------------------------

# Convert first 3 letters to uppercase, rest to lowercase
letters_list = (
    [letter.upper() for letter in letters_list[:3]] +
    [letter.lower() for letter in letters_list[3:]]
)

# Reconstruct the name into a string
reconstructed_name = "".join(letters_list)
print("Reconstructed name:", reconstructed_name)

# -----------------------------
# Simple Arithmetic on a List
# -----------------------------

# Prompt the user to enter 5 numbers
user_input = input("Enter 5 numbers separated by spaces: ")

# Convert input to a list of integers
numbers_list = [int(num) for num in user_input.split()]

# Calculate sum and average
total = sum(numbers_list)
average = total / len(numbers_list)

# Print results
print("Numbers list:", numbers_list)
print("Sum:", total)
print("Average:", average)
```
